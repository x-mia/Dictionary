```
from pandas import read_csv

df = read_csv("eesti-slovaki-sonastik.csv")
```

```
def replacing(row):

  link = row["Sõnaveeb"]
  est_word = row["Eesti sõna"]
  if link != "-":
    return "https://sonaveeb.ee/search/unif/dlall/dsall/"+ est_word +"/1"
  if link == "-":
    return "-"



df["Sõnaveeb"] = df.apply(replacing, axis=1)

  
```

```
df.to_csv("eesti-slovaki-sonastik.csv", index=False)
```
